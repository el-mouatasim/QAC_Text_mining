### PERF فعل ماض Perfect verb

```python
qdf_perf=pd.read_sql("select * from QACV_TABLE where Aspect='PERF'", conn)
with pd.option_context('display.max_colwidth', None):
  display(qdf_perf.head())
```

```python
list1=qdf_perf['Voice'].values.tolist()
Voice_co=pd.Series(list1).value_counts(); Voice_co
```

```python
list2=qdf_perf['MOOD'].values.tolist()
MOOD_co=pd.Series(list2).value_counts(); MOOD_co
```

```python
list3=qdf_perf['PGN'].values.tolist()
PGN_co=pd.Series(list3).value_counts(); PGN_co.keys()
```

```python
#rest ['3MP', '1P', '2MP', '3FS', '2MS', '1S', '3MD', '3FP', '3FD','2FP', '2FS', '2D']
def fetcol(aspect,voice,pgn,mood='IND'):
    dfverb=pd.read_sql(f"""select * from QACV_TABLE where Aspect = '{aspect}' and Voice= '{voice}' and PGN = '{pgn}' and MOOD ='{mood}' """, conn)
    dfverb_d = pd.DataFrame({'Root': list(dfverb['RootAr']),'FormV': list(dfverb['FormV'])})
    dfverb_d = dfverb_d.groupby(['Root', 'FormV']).size().reset_index().rename(columns={0: '{}_{}_{}_{}'.format(aspect,voice,mood,pgn)})
    return dfverb_d
```

```python
list_pgn= ['3MP', '1P', '2MP', '3FS', '2MS', '1S', '3MD', '3FP', '3FD','2FP', '2FS', '2D']
perf_act=fetcol('PERF','ACT','3MS')
for pgn in list_pgn:
    perf_act2=fetcol('PERF','ACT',pgn)
    #combine  perf_act and perf_act2
    perf_act = perf_act.merge(perf_act2, on=['Root',  'FormV'], how='left')
perf_act
```

```python
list_pgn= ['3MP', '1P', '2MP', '3FS', '2MS', '1S', '3MD', '3FP', '3FD','2FP', '2FS', '2D']
perf_pass=fetcol('PERF','PASS','3MS')
for pgn in list_pgn:
    perf_pass2=fetcol('PERF','PASS',pgn)
    #combine  perf_pass and perf_pass2
    perf_pass = perf_pass.merge(perf_pass2, on=['Root',  'FormV'], how='left')
perf_pass
```

```python
#combine  perf_act and perf_pass
perf_all = perf_act.merge(perf_pass, on=['Root',  'FormV'], how='left'); perf_all
```

### IMPF فعل مضارع Imperfect verb


```python
qdf_impf=pd.read_sql("select * from QACV_TABLE where Aspect='IMPF'", conn)
with pd.option_context('display.max_colwidth', None):
  display(qdf_impf.head())
```

```python
list1=qdf_impf['Voice'].values.tolist()
Voice_co=pd.Series(list1).value_counts(); Voice_co
```

```python
list2=qdf_impf['MOOD'].values.tolist()
MOOD_co=pd.Series(list2).value_counts(); MOOD_co
```

```python
list3=qdf_impf['PGN'].values.tolist()
PGN_co=pd.Series(list3).value_counts(); PGN_co.keys()
```

```python
list_pgn= ['3MP', '2MP', '1P', '3FS', '2MS', '1S', '3FP', '2D', '3MD', '2FS', '2FP', '2FD', '2MD', '3FD']
list_mood = ['IND','JUS','SUBJ']
impf_act=fetcol('IMPF','ACT','3MS')
for pgn in list_pgn:
    for mood in list_mood:
        impf_act2=fetcol('IMPF','ACT',pgn, mood)
        #combine  impf_act and impf_act2
        impf_act = impf_act.merge(impf_act2, on=['Root',  'FormV'], how='left')
impf_act
```

```python
list_pgn= ['3MP', '2MP', '1P', '3FS', '2MS', '1S', '3FP', '2D', '3MD', '2FS', '2FP', '2FD', '2MD', '3FD']
list_mood = ['IND','JUS','SUBJ']
impf_pass=fetcol('IMPF','PASS','3MS')
for pgn in list_pgn:
    for mood in list_mood:
        impf_pass2=fetcol('IMPF','PASS',pgn,mood)
        #combine  impf_pass and impf_pass2
        impf_pass = impf_pass.merge(impf_pass2, on=['Root',  'FormV'], how='left')
impf_pass
```

```python
#combine  impf_act and impf_pass
impf_all = impf_act.merge(impf_pass, on=['Root',  'FormV'], how='left'); impf_all
```

## Derivation features
ACT PCPL 	اسم فاعل 	Active participle <br>
PASS PCPL 	اسم مفعول 	Passive participle<br>
VN 	مصدر 	Verbal noun

```python
qdf_n=pd.read_sql("select * from QACN_TABLE", conn)
with pd.option_context('display.max_colwidth', None):
  display(qdf_n.head())
```

```python
list1=qdf_n['Derivation'].values.tolist()
Derivation_co=pd.Series(list1).value_counts(); Derivation_co
```

```python
list1=qdf_n['State'].values.tolist()
State_co=pd.Series(list1).value_counts(); State_co
```

```python
list1=qdf_n['Grop'].values.tolist()
Grop_co=pd.Series(list1).value_counts(); Grop_co.keys()
```

```python
def nfetcol(Derivation,State,Grop):
    dfn=pd.read_sql(f"""select * from QACN_TABLE where Derivation = '{Derivation}' and State= '{State}' and Grop = '{Grop}'  """, conn)
    dfn_d = pd.DataFrame({'Root': list(dfn['RootAr']),'FormN': list(dfn['FormN'])})
    dfn_d = dfn_d.groupby(['Root', 'FormN']).size().reset_index().rename(columns={0: 'N_{}_{}_{}'.format(Derivation,State,Grop)})
    return dfn_d
```

```python
list_pgn= ['MP','', 'M', 'FP', 'F', 'MD', 'FD', 'MS', 'FS', 'P']
list_deriv=['ACT PCPL','PASS PCPL','VN']
list_stat=['DEF','INDEF']
star=0
for der in list_deriv:
    for st in list_stat:
        for pgn in list_pgn:
            if star==0:
                name_all=nfetcol(der,st,pgn)
                star+=1
            else:
                name2=nfetcol(der,st,pgn)
                name_all = name_all.merge(name2, on=['Root',  'FormN'], how='left')
name_all
```

```python
def afetcol(Derivation,State,Grop):
    dfn=pd.read_sql(f"""select * from QACA_TABLE where Derivation = '{Derivation}' and State= '{State}' and Grop = '{Grop}'  """, conn)
    dfn_d = pd.DataFrame({'Root': list(dfn['RootAr']),'FormN': list(dfn['FormN'])})
    dfn_d = dfn_d.groupby(['Root', 'FormN']).size().reset_index().rename(columns={0: 'Adj_{}_{}_{}'.format(Derivation,State,Grop)})
    return dfn_d
```

```python
list_pgn= ['MP','', 'M', 'FP', 'F', 'MD', 'FD', 'MS', 'FS', 'P']
list_deriv=['ACT PCPL','PASS PCPL','VN']
list_stat=['DEF','INDEF']
star=0
for der in list_deriv:
    for st in list_stat:
        for pgn in list_pgn:
            if star==0:
                adj_all=afetcol(der,st,pgn)
                star+=1
            else:
                name2=afetcol(der,st,pgn)
                adj_all = adj_all.merge(name2, on=['Root',  'FormN'], how='left')
adj_all
```

```python
name_adj_all = name_all.merge(adj_all, on=['Root',  'FormN'], how='left');name_adj_all
```

```python
name_adj_all.to_csv(r'name_adj_root_all.csv', index=False,header=True, encoding='utf-8-sig')
```
